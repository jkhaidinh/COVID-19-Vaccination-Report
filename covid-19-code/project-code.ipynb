```python
# libraries
import pandas as pd
import altair as alt
```

```python
# tidying poli
poli = pd.read_excel('county.xlsx')
poli

politidy = poli.dropna(how = 'any')
politidy = politidy[['County', 'Eligible', 'Total Registered', 'Democratic','Republican']]

politidy = politidy.astype({'Eligible' : 'int', 'Total Registered' : 'int', 'Democratic' : 'int', 'Republican' : 'int'})
politidy = politidy.reset_index().drop(columns = ['index']).drop(58,axis=0)
politidy["Democratic Percent"] = politidy.loc[:,"Democratic"]/politidy.loc[:, "Total Registered"]
politidy["Republican Percent"] = politidy.loc[:,"Republican"]/politidy.loc[:, "Total Registered"]
politidy.head()

#politidy.to_csv('poli-tidy.csv', index = False)
```

```python
# tidying vax
vax = pd.read_csv("covid-19-vaccines-by-zip-code-5-12-21.csv")

vaxtidy = vax.dropna(how = 'any')

vaxtidy = vaxtidy[['Zip Code Tabulation Area', 'County', 'Vaccine Equity Metric Quartile', '16+ Population', 'Persons Fully Vaccinated',
                   'Persons Partially Vaccinated', 'Percent of Population with 1+ Dose']].reset_index().drop(columns = 'index')
vaxtidy.head()

#vaxtidy.to_csv('vax-tidy.csv', index = False)
```

```python
# Do different VEM scores show different percentages of vaccinations?
vem_1d_1 = vaxtidy[vaxtidy['Vaccine Equity Metric Quartile'] == 1].loc[:, 'Percent of Population with 1+ Dose'].mean()

vem_1d_2 = vaxtidy[vaxtidy['Vaccine Equity Metric Quartile'] == 2].loc[:, 'Percent of Population with 1+ Dose'].mean()

vem_1d_3 = vaxtidy[vaxtidy['Vaccine Equity Metric Quartile'] == 3].loc[:, 'Percent of Population with 1+ Dose'].mean()

vem_1d_4 = vaxtidy[vaxtidy['Vaccine Equity Metric Quartile'] == 4].loc[:, 'Percent of Population with 1+ Dose'].mean()

vemunique = vaxtidy['Vaccine Equity Metric Quartile'].sort_values(ascending = True).unique()

vemavg = vaxtidy[]

data = {'Vaccine Equity Metric Quartile' : vemunique, 'Percent of Population with 1+ Dose' : [vem_1d_1, vem_1d_2, vem_1d_3, vem_1d_4]}

table1 = pd.DataFrame(data = data)
table1

vemchart = alt.Chart(table1).mark_bar(size = 50).encode(
    x = alt.X('Vaccine Equity Metric Quartile',
             axis = alt.Axis(tickCount = 4)),
    y = alt.Y('Percent of Population with 1+ Dose',
             scale = alt.Scale(domain = (0,1))),
    color = alt.Color('Vaccine Equity Metric Quartile', legend = None)
)

vemchart
```
